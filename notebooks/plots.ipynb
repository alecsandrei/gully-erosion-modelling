```python
import seaborn as sns
import matplotlib.pyplot as plt
import geopandas as gpd
import pathlib
```

```python
data = gpd.read_file(
    '/media/alex/My Passport/FACULTATE_FOLDERE/STUDII/RAVENE_2025/qgs_interim_files/Profile_soldanesti_amonte.shp'
).rename(
    columns={
        'gully_cover': 'gully cover',
        'estimated D': 'estimated topography',
        'soldanesti_': 'ground truth',
    }
)[['Distance', 'estimated topography', 'ground truth', 'gully cover']]

fig, ax = plt.subplots(figsize=(10, 5))
for y, color in zip(
    ['estimated topography', 'ground truth', 'gully cover'],
    ['#33ff2c', 'blue', 'black'],
):
    sns.lineplot(data, x='Distance', y=y, c=color, ax=ax, label=y, linewidth=3)
ax.legend(prop={'size': 12})
plt.ylim(176.5, 181)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylabel('Elevation (m)', size=15)
plt.xlabel('Distance', size=15)
plt.tight_layout()
plt.savefig(
    r'/media/alex/My Passport/FACULTATE_FOLDERE/STUDII/RAVENE_2025/profile.png',
    dpi=300,
)
```

```python
for shp in pathlib.Path(
    '/media/alex/My Passport/FACULTATE_FOLDERE/STUDII/RAVENE_2025/qgs_interim_files/soldanesti_aval_profiles'
).glob('*.shp'):
    data = gpd.read_file(shp).rename(
        columns={
            'GULLYCOVER': 'gully cover',
            'INTERPOLATE': 'estimated DEM',
            'soldanestia': 'ground truth',
        }
    )[['DISTANCE', 'estimated DEM', 'ground truth', 'gully cover']]

    fig, ax = plt.subplots(figsize=(8, 5))
    for y, color in zip(
        ['estimated DEM', 'ground truth', 'gully cover'],
        ['#33ff2c', 'blue', 'black'],
    ):
        sns.lineplot(data, x='DISTANCE', y=y, c=color, ax=ax, label=y, linewidth=5)
    ax.legend().remove()
    # plt.ylim(176.5, 181)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.ylabel('Elevation (m)', size=15)
    plt.xlabel('Distance', size=15)
    plt.tight_layout()
    plt.savefig(
        f'/media/alex/My Passport/FACULTATE_FOLDERE/STUDII/RAVENE_2025/reports{shp.stem}.png',
        dpi=300,
    )
```
